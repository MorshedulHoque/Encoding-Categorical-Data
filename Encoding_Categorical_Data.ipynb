```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/Datasets/customer.csv')
```

```python
df.sample(5)
```

```python
df = df[['review', 'education', 'purchased']]
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df[['review', 'education']], df['purchased'], test_size=0.3)
```

```python
X_train
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']])
```

```python
oe.fit(X_train)
```

```python
X_train = oe.transform(X_train)
```

```python
X_train
```

```python
oe.categories_
```

```python
X_train
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
le.fit(y_train)
```

```python
le.classes_
```

```python
y_train = le.transform(y_train)
y_test = le.transform(y_test)
```
